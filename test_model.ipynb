```python
# 📦 Install required libraries
!pip install yfinance joblib pandas scikit-learn --quiet

# 📚 Import packages
import yfinance as yf
import pandas as pd
import joblib

# 📥 Load trained model
model = joblib.load('/content/drive/MyDrive/ml_stocks/model/simple_stock_model.pkl')

# 📊 Function to compute features
def compute_features(df):
    df['SMA_5'] = df['Close'].rolling(window=5).mean()
    df['SMA_10'] = df['Close'].rolling(window=10).mean()
    df['Daily_Return'] = df['Close'].pct_change()
    df['Volatility'] = df['Close'].rolling(window=5).std()
    df['Volume_Change'] = df['Volume'].pct_change()
    df.dropna(inplace=True)
    return df[['SMA_5', 'SMA_10', 'Daily_Return', 'Volatility', 'Volume_Change']]

# 🧪 Input: any valid stock symbol (e.g., RELIANCE.NS)
stock = input("Enter NIFTY50 stock symbol (e.g., RELIANCE.NS): ").strip()

# 📈 Download test data
df = yf.download(stock, period='3mo', interval='1d')
if df.empty:
    print("❌ Invalid or no data for stock symbol.")
else:
    # 🔍 Extract latest features
    X_test = compute_features(df.copy())
    latest = X_test.iloc[-1].values.reshape(1, -1)

    # 🎯 Predict
    prediction = model.predict(latest)[0]

    print("\n✅ Prediction for next day:")
    if prediction == 1:
        print("📈 Stock is likely to RISE – Consider BUY")
    else:
        print("📉 Stock is likely to FALL – Consider SELL")
```
